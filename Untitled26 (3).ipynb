```python
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(42)
```

```python
# ─────────────────────────────────────────────
#  CONFIGURATION — edit these to tune
# ─────────────────────────────────────────────
CONFIG = {
    "img_h"        : 512,
    "img_w"        : 512,
    "batch_size"   : 2,       # safe for RTX 4070 8 GB at 512x512
    "n_epochs"     : 50,
    "lr"           : 6e-5,    # small LR for fine-tuning
    "weight_decay" : 0.01,
    "patience"     : 10,      # early stop patience

    # SegFormer variant:
    #   "nvidia/mit-b0"  — lightest,  fastest  (~4 M params)
    #   "nvidia/mit-b2"  — balanced,  recommended  (~25 M params)
    #   "nvidia/mit-b4"  — strongest, needs more VRAM (~65 M params)
    "model_name"   : "nvidia/mit-b2",

    # Class weights — higher = model penalised more for missing that class
    # Order: BG, Trees, LushBush, DryGrass, DryBush,
    #        GndClutter, Logs, Rocks, Landscape, Sky
    "class_weights": [
        0.5,   # 0  Background
        1.0,   # 1  Trees
        1.5,   # 2  Lush Bushes
        1.0,   # 3  Dry Grass
        1.5,   # 4  Dry Bushes
        3.0,   # 5  Ground Clutter  ← rare
        4.0,   # 6  Logs            ← very rare
        2.0,   # 7  Rocks
        0.5,   # 8  Landscape       ← dominant
        0.3,   # 9  Sky             ← dominant
    ],
}
```

```python
def convert_mask(mask: Image.Image) -> np.ndarray:
    arr = np.array(mask)
    out = np.zeros_like(arr, dtype=np.uint8)
    for raw, cls in VALUE_MAP.items():
        out[arr == raw] = cls
    return out
```

```python
# ─────────────────────────────────────────────
#  JOINT AUGMENTATION
# ─────────────────────────────────────────────
class JointAugment:
    """Same spatial transform applied to image AND mask."""
    def __init__(self, img_h, img_w, train=True):
        self.h, self.w = img_h, img_w
        self.train     = train

    def __call__(self, image: Image.Image, mask_np: np.ndarray):
        # Always resize
        image    = TF.resize(image,
                             (self.h, self.w), interpolation=Image.BILINEAR)
        mask_pil = Image.fromarray(mask_np)
        mask_pil = TF.resize(mask_pil,
                             (self.h, self.w), interpolation=Image.NEAREST)

        if self.train:
            # Horizontal flip 50 %
            if random.random() > 0.5:
                image    = TF.hflip(image)
                mask_pil = TF.hflip(mask_pil)

            # Vertical flip 20 %
            if random.random() > 0.8:
                image    = TF.vflip(image)
                mask_pil = TF.vflip(mask_pil)

            # Random rotation ±15°
            if random.random() > 0.5:
                angle    = random.uniform(-15, 15)
                image    = TF.rotate(image,    angle,
                                     interpolation=Image.BILINEAR)
                mask_pil = TF.rotate(mask_pil, angle,
                                     interpolation=Image.NEAREST)

            # Random crop + resize (scale 0.7–1.0)
            if random.random() > 0.4:
                scale  = random.uniform(0.7, 1.0)
                crop_h = int(self.h * scale)
                crop_w = int(self.w * scale)
                max_i  = max(self.h - crop_h, 1)
                max_j  = max(self.w - crop_w, 1)
                i = random.randint(0, max_i)
                j = random.randint(0, max_j)
                image    = TF.crop(image,    i, j, crop_h, crop_w)
                mask_pil = TF.crop(mask_pil, i, j, crop_h, crop_w)
                image    = TF.resize(image,
                                     (self.h, self.w), interpolation=Image.BILINEAR)
                mask_pil = TF.resize(mask_pil,
                                     (self.h, self.w), interpolation=Image.NEAREST)

            # Colour jitter 70 %
            if random.random() > 0.3:
                image = TF.adjust_brightness(image, random.uniform(0.7, 1.3))
                image = TF.adjust_contrast  (image, random.uniform(0.7, 1.3))
                image = TF.adjust_saturation(image, random.uniform(0.7, 1.3))
                image = TF.adjust_hue       (image, random.uniform(-0.08, 0.08))

        image = TF.to_tensor(image)
        image = TF.normalize(image,
                             mean=[0.485, 0.456, 0.406],
                             std =[0.229, 0.224, 0.225])
        mask  = torch.from_numpy(np.array(mask_pil)).long()
        return image, mask
```

```python
class SegDataset(Dataset):
    def __init__(self, data_dir: str, augment: JointAugment):
        self.image_dir = os.path.join(data_dir, 'Color_Images')
        self.masks_dir = os.path.join(data_dir, 'Segmentation')
        self.augment   = augment

        self.ids = sorted([
            f for f in os.listdir(self.image_dir)
            if os.path.isfile(os.path.join(self.image_dir, f))
            and f.endswith(('.png', '.jpg', '.jpeg'))
        ])

    def __len__(self):
        return len(self.ids)

    def __getitem__(self, idx):
        name = self.ids[idx]

        image = Image.open(os.path.join(self.image_dir, name)).convert('RGB')
        mask  = Image.open(os.path.join(self.masks_dir, name))

        mask_np = convert_mask(mask)

        image, mask = self.augment(image, mask_np)

        return image, mask
```

```python
# ─────────────────────────────────────────────
#  METRICS
# ─────────────────────────────────────────────
def compute_iou(logits, target, num_classes=N_CLASSES):
    pred = logits.argmax(dim=1).view(-1)
    tgt  = target.view(-1)
    per  = []
    for c in range(num_classes):
        p = pred == c;  t = tgt == c
        inter = (p & t).sum().float()
        union = (p | t).sum().float()
        per.append(float('nan') if union == 0
                   else (inter / union).item())
    return float(np.nanmean(per)), per


def compute_dice(logits, target, num_classes=N_CLASSES, smooth=1e-6):
    pred = logits.argmax(dim=1).view(-1)
    tgt  = target.view(-1)
    sc   = []
    for c in range(num_classes):
        p = (pred == c).float();  t = (tgt == c).float()
        sc.append(((2*(p*t).sum() + smooth) /
                   (p.sum() + t.sum() + smooth)).item())
    return float(np.mean(sc))


def compute_pixel_acc(logits, target):
    return (logits.argmax(dim=1) == target).float().mean().item()


@torch.no_grad()
def evaluate(model, loader, device, criterion):
    model.eval()
    iou_l, dice_l, acc_l, loss_l = [], [], [], []
    for imgs, labels in loader:
        imgs, labels = imgs.to(device), labels.to(device)
        out    = model(pixel_values=imgs)
        logits = F.interpolate(out.logits, size=imgs.shape[2:],
                               mode='bilinear', align_corners=False)
        loss   = criterion(logits, labels)
        iou, _ = compute_iou(logits, labels)
        iou_l.append(iou);   dice_l.append(compute_dice(logits, labels))
        acc_l.append(compute_pixel_acc(logits, labels))
        loss_l.append(loss.item())
    model.train()
    return (float(np.nanmean(iou_l)), float(np.mean(dice_l)),
            float(np.mean(acc_l)),    float(np.mean(loss_l)))
```

```python
# ─────────────────────────────────────────────
#  PLOTS + LOG
# ─────────────────────────────────────────────
def save_plots(history, out_dir):
    os.makedirs(out_dir, exist_ok=True)
    epochs = range(1, len(history['train_loss']) + 1)
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    pairs = [
        ('train_loss','val_loss','Loss','Loss'),
        ('train_iou', 'val_iou', 'IoU', 'IoU'),
        ('train_dice','val_dice','Dice Score','Dice'),
        ('train_pixel_acc','val_pixel_acc','Pixel Accuracy','Accuracy'),
    ]
    for ax, (tk, vk, title, ylabel) in zip(axes.flat, pairs):
        ax.plot(epochs, history[tk], 'o-', label='Train', linewidth=2)
        ax.plot(epochs, history[vk], 's-', label='Val',   linewidth=2)
        ax.set_title(title, fontsize=13, fontweight='bold')
        ax.set_xlabel('Epoch'); ax.set_ylabel(ylabel)
        ax.legend(); ax.grid(True, alpha=0.3)
    plt.tight_layout()
    p = os.path.join(out_dir, 'all_metrics.png')
    plt.savefig(p, dpi=150); plt.close()
    print(f"  Saved → {p}")


def save_per_class_plot(per_class, out_dir, epoch):
    os.makedirs(out_dir, exist_ok=True)
    vals   = [x if not np.isnan(x) else 0 for x in per_class]
    colors = ['#2ecc71' if v >= 0.5 else '#e67e22' if v >= 0.3
              else '#e74c3c' for v in vals]
    fig, ax = plt.subplots(figsize=(12, 5))
    ax.bar(CLASS_NAMES, vals, color=colors, edgecolor='white')
    mean_v = float(np.nanmean(per_class))
    ax.axhline(mean_v, color='red', linestyle='--', linewidth=1.5,
               label=f'Mean IoU = {mean_v:.3f}')
    ax.set_ylim(0, 1); ax.set_ylabel('IoU')
    ax.set_title(f'Per-Class IoU — Epoch {epoch}', fontweight='bold')
    ax.legend(); ax.grid(axis='y', alpha=0.3)
    plt.xticks(rotation=35, ha='right'); plt.tight_layout()
    p = os.path.join(out_dir, f'per_class_iou_epoch{epoch:02d}.png')
    plt.savefig(p, dpi=150); plt.close()


def save_log(history, best_epoch, out_dir):
    os.makedirs(out_dir, exist_ok=True)
    p = os.path.join(out_dir, 'training_log.txt')
    with open(p, 'w') as f:
        f.write("=" * 80 + "\n")
        f.write("  DUALITY AI HACKATHON — SEGFORMER-B2 TRAINING LOG\n")
        f.write("=" * 80 + "\n\n")
        f.write(f"  Model         : {CONFIG['model_name']}\n")
        f.write(f"  Image size    : {CONFIG['img_h']} x {CONFIG['img_w']}\n")
        f.write(f"  Batch size    : {CONFIG['batch_size']}\n")
        f.write(f"  Learning rate : {CONFIG['lr']}\n\n")
        f.write(f"  Best Val IoU  : {max(history['val_iou']):.4f}  (Epoch {best_epoch})\n")
        f.write(f"  Best Val Dice : {max(history['val_dice']):.4f}\n")
        f.write(f"  Best Val Acc  : {max(history['val_pixel_acc']):.4f}\n\n")
        hdr = ('Epoch','TrLoss','VaLoss','TrIoU','VaIoU',
               'TrDice','VaDice','TrAcc','VaAcc')
        f.write("{:<7}{:<10}{:<10}{:<10}{:<10}"
                "{:<10}{:<10}{:<10}{:<10}\n".format(*hdr))
        f.write("-" * 80 + "\n")
        for i in range(len(history['train_loss'])):
            mark = " ★" if (i + 1) == best_epoch else ""
            f.write("{:<7}{:<10.4f}{:<10.4f}{:<10.4f}{:<10.4f}"
                    "{:<10.4f}{:<10.4f}{:<10.4f}{:<10.4f}{}\n".format(
                i+1,
                history['train_loss'][i], history['val_loss'][i],
                history['train_iou'][i],  history['val_iou'][i],
                history['train_dice'][i], history['val_dice'][i],
                history['train_pixel_acc'][i], history['val_pixel_acc'][i],
                mark))
    print(f"  Saved → {p}")
```

```python
# ─────────────────────────────────────────────
#  MAIN
# ─────────────────────────────────────────────
def main():
# ====== COLAB PATH SETUP ======
    base_dir = "/content/drive/MyDrive/Offroad_Segmentation_Training_Dataset/Offroad_Segmentation_Training_Dataset"   # 🔁 change if needed

    train_dir = os.path.join(base_dir, "train")
    val_dir   = os.path.join(base_dir, "val")

    out_dir   = os.path.join(base_dir, "outputs")
    last_path = os.path.join(base_dir, "segformer_last.pth")
    best_path = os.path.join(base_dir, "segformer_best.pth")

    os.makedirs(out_dir, exist_ok=True)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"\n{'='*60}")
    print(f"  Device        : {device}")
    if device.type == 'cuda':
        print(f"  GPU           : {torch.cuda.get_device_name(0)}")
        vram = torch.cuda.get_device_properties(0).total_memory / 1e9
        print(f"  VRAM          : {vram:.1f} GB")
    print(f"  Model         : {CONFIG['model_name']}")
    print(f"  Image size    : {CONFIG['img_h']} x {CONFIG['img_w']}")
    print(f"  Batch size    : {CONFIG['batch_size']}")
    print(f"  Epochs        : {CONFIG['n_epochs']}")
    print(f"  LR            : {CONFIG['lr']}")
    print(f"{'='*60}\n")

    # ── Datasets ─────────────────────────────
    train_aug = JointAugment(CONFIG['img_h'], CONFIG['img_w'], train=True)
    val_aug   = JointAugment(CONFIG['img_h'], CONFIG['img_w'], train=False)
    train_set = SegDataset(train_dir, train_aug)
    val_set   = SegDataset(val_dir,   val_aug)
    train_loader = DataLoader(train_set, batch_size=CONFIG['batch_size'],
                              shuffle=True,  num_workers=2, pin_memory=True)
    val_loader   = DataLoader(val_set,   batch_size=CONFIG['batch_size'],
                              shuffle=False, num_workers=2, pin_memory=True)
    print(f"  Train samples : {len(train_set)}")
    print(f"  Val   samples : {len(val_set)}\n")

    # ── Load SegFormer-B2 ────────────────────
    print(f"  Loading {CONFIG['model_name']} ...")
    print("  (First run downloads ~350 MB — cached after that)\n")
    model = SegformerForSemanticSegmentation.from_pretrained(
        CONFIG['model_name'],
        num_labels=N_CLASSES,
        ignore_mismatched_sizes=True,
        id2label={i: n for i, n in enumerate(CLASS_NAMES)},
        label2id={n: i for i, n in enumerate(CLASS_NAMES)},
    ).to(device)

    tp = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"  Trainable params : {tp:,}\n")

    # ── Loss ─────────────────────────────────
    weights   = torch.tensor(CONFIG['class_weights'], dtype=torch.float).to(device)
    criterion = nn.CrossEntropyLoss(weight=weights)

    # ── Optimiser — encoder gets 10x smaller LR ──
    optimizer = optim.AdamW([
        {'params': model.segformer.parameters(),   'lr': CONFIG['lr'] * 0.1},
        {'params': model.decode_head.parameters(), 'lr': CONFIG['lr']},
    ], weight_decay=CONFIG['weight_decay'])

    scheduler = optim.lr_scheduler.CosineAnnealingLR(
        optimizer, T_max=CONFIG['n_epochs'], eta_min=1e-7)

    use_amp = device.type == 'cuda'
    scaler  = torch.cuda.amp.GradScaler() if use_amp else None

    history = {k: [] for k in [
        'train_loss','val_loss','train_iou','val_iou',
        'train_dice','val_dice','train_pixel_acc','val_pixel_acc']}

    best_val_iou, best_epoch, patience_count = 0.0, 1, 0

    print("=" * 60)
    print("  TRAINING START")
    print("=" * 60)

    for epoch in range(1, CONFIG['n_epochs'] + 1):
        model.train()
        train_losses = []
        pbar = tqdm(train_loader,
                    desc=f"Epoch {epoch:02d}/{CONFIG['n_epochs']} [Train]",
                    leave=False)

        for imgs, labels in pbar:
            imgs, labels = imgs.to(device), labels.to(device)
            optimizer.zero_grad(set_to_none=True)

            if use_amp:
                with torch.cuda.amp.autocast():
                    out    = model(pixel_values=imgs)
                    logits = F.interpolate(out.logits, size=imgs.shape[2:],
                                           mode='bilinear', align_corners=False)
                    loss   = criterion(logits, labels)
                scaler.scale(loss).backward()
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                scaler.step(optimizer)
                scaler.update()
            else:
                out    = model(pixel_values=imgs)
                logits = F.interpolate(out.logits, size=imgs.shape[2:],
                                       mode='bilinear', align_corners=False)
                loss   = criterion(logits, labels)
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()

            train_losses.append(loss.item())
            pbar.set_postfix(loss=f"{loss.item():.4f}")

        scheduler.step()

        # ── Metrics ──────────────────────────
        val_iou,   val_dice,   val_acc,   val_loss   = evaluate(model, val_loader,   device, criterion)
        train_iou, train_dice, train_acc, _           = evaluate(model, train_loader, device, criterion)
        t_loss = float(np.mean(train_losses))

        for k, v in zip(
            ['train_loss','val_loss','train_iou','val_iou',
             'train_dice','val_dice','train_pixel_acc','val_pixel_acc'],
            [t_loss, val_loss, train_iou, val_iou,
             train_dice, val_dice, train_acc, val_acc]
        ):
            history[k].append(v)

        # ── Per-class IoU (val) ───────────────
        model.eval()
        accum = [[] for _ in range(N_CLASSES)]
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs, labels = imgs.to(device), labels.to(device)
                out    = model(pixel_values=imgs)
                logits = F.interpolate(out.logits, size=imgs.shape[2:],
                                       mode='bilinear', align_corners=False)
                _, pc  = compute_iou(logits, labels)
                for c in range(N_CLASSES):
                    if not np.isnan(pc[c]):
                        accum[c].append(pc[c])
        avg_pc = [float(np.mean(v)) if v else float('nan') for v in accum]
        model.train()

        # ── Console output ────────────────────
        lr_d = optimizer.param_groups[1]['lr']
        print(f"\nEpoch {epoch:02d}/{CONFIG['n_epochs']}  lr={lr_d:.1e}  |  "
              f"T-loss={t_loss:.4f} V-loss={val_loss:.4f}  |  "
              f"V-IoU={val_iou:.4f} V-Dice={val_dice:.4f} V-Acc={val_acc:.4f}")
        print("  Per-class IoU :")
        for name, iou in zip(CLASS_NAMES, avg_pc):
            flag = " ← LOW" if (not np.isnan(iou) and iou < 0.3) else ""
            bar  = "█" * int((iou if not np.isnan(iou) else 0) * 20)
            print(f"    {name:<20} {iou:.3f}  {bar}{flag}")

        if epoch % 5 == 0 or epoch == CONFIG['n_epochs']:
            save_per_class_plot(avg_pc, out_dir, epoch)

        # ── Checkpoint ───────────────────────
        if val_iou > best_val_iou:
            best_val_iou, best_epoch, patience_count = val_iou, epoch, 0
            torch.save(model.state_dict(), best_path)
            print(f"  ★  New best Val IoU = {best_val_iou:.4f} → saved")
        else:
            patience_count += 1
            print(f"  No improvement ({patience_count}/{CONFIG['patience']})")

        if patience_count >= CONFIG['patience']:
            print(f"\n  Early stopping at epoch {epoch}.")
            break

    torch.save(model.state_dict(), last_path)
    save_plots(history, out_dir)
    save_log(history, best_epoch, out_dir)

    print("\n" + "=" * 60)
    print("  TRAINING COMPLETE")
    print("=" * 60)
    print(f"  Best Val IoU   : {best_val_iou:.4f}  (Epoch {best_epoch})")
    print(f"  Best Val Dice  : {max(history['val_dice']):.4f}")
    print(f"  Best Val Acc   : {max(history['val_pixel_acc']):.4f}")
    print(f"  Best model     : {best_path}")
    print(f"  Stats          : {out_dir}")
    print("=" * 60)
    print("\n  Next step → run:  python val.py")


if __name__ == "__main__":
    main()
```

```python
"""
=============================================================================
  Duality AI Hackathon — SegFormer-B2 Training (Google Colab)
  Run this entire file as a single Colab cell or .py script

  Before running, execute in a separate cell:
      !pip install transformers -q
      from google.colab import drive
      drive.mount('/content/drive')
=============================================================================
"""

# ── Imports ───────────────────────────────────────────────────────────────
import os
import random
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision.transforms.functional as TF
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from PIL import Image
from tqdm import tqdm
from transformers import SegformerForSemanticSegmentation
import matplotlib
matplotlib.use('Agg')
plt.switch_backend('Agg')


# ─────────────────────────────────────────────
#  REPRODUCIBILITY
# ─────────────────────────────────────────────
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(42)


# ─────────────────────────────────────────────
#  CONFIGURATION
# ─────────────────────────────────────────────
CONFIG = {
    "img_h"        : 512,
    "img_w"        : 512,
    "batch_size"   : 4,       # T4 15 GB — safe at 512x512 batch 4
    "n_epochs"     : 50,
    "lr"           : 6e-5,
    "weight_decay" : 0.01,
    "patience"     : 10,
    "model_name"   : "nvidia/mit-b2",

    # Higher weight = model penalised more for missing that class
    # Rare classes (Logs, Ground Clutter) get highest weights
    "class_weights": [
        0.5,   # 0  Background
        1.0,   # 1  Trees
        1.5,   # 2  Lush Bushes
        1.0,   # 3  Dry Grass
        1.5,   # 4  Dry Bushes
        3.0,   # 5  Ground Clutter  ← rare
        4.0,   # 6  Logs            ← very rare
        2.0,   # 7  Rocks
        0.5,   # 8  Landscape       ← dominant
        0.3,   # 9  Sky             ← dominant
    ],
}


# ─────────────────────────────────────────────
#  CLASS DEFINITIONS
# ─────────────────────────────────────────────
VALUE_MAP = {
    0:     0,
    100:   1,
    200:   2,
    300:   3,
    500:   4,
    550:   5,
    700:   6,
    800:   7,
    7100:  8,
    10000: 9,
}
N_CLASSES   = len(VALUE_MAP)
CLASS_NAMES = [
    'Background', 'Trees', 'Lush Bushes', 'Dry Grass', 'Dry Bushes',
    'Ground Clutter', 'Logs', 'Rocks', 'Landscape', 'Sky'
]


# ─────────────────────────────────────────────
#  MASK CONVERSION
# ─────────────────────────────────────────────
def convert_mask(mask: Image.Image) -> np.ndarray:
    """Raw pixel values → class IDs 0-9."""
    arr = np.array(mask)
    out = np.zeros_like(arr, dtype=np.uint8)
    for raw, cls in VALUE_MAP.items():
        out[arr == raw] = cls
    return out


# ─────────────────────────────────────────────
#  JOINT AUGMENTATION
# ─────────────────────────────────────────────
class JointAugment:
    """Same spatial transform on image AND mask."""
    def __init__(self, img_h, img_w, train=True):
        self.h, self.w = img_h, img_w
        self.train     = train

    def __call__(self, image: Image.Image, mask_np: np.ndarray):
        # Resize both
        image    = TF.resize(image,
                             (self.h, self.w), interpolation=Image.BILINEAR)
        mask_pil = Image.fromarray(mask_np)
        mask_pil = TF.resize(mask_pil,
                             (self.h, self.w), interpolation=Image.NEAREST)

        if self.train:
            # Horizontal flip 50%
            if random.random() > 0.5:
                image    = TF.hflip(image)
                mask_pil = TF.hflip(mask_pil)

            # Vertical flip 20%
            if random.random() > 0.8:
                image    = TF.vflip(image)
                mask_pil = TF.vflip(mask_pil)

            # Random rotation ±15°
            if random.random() > 0.5:
                angle    = random.uniform(-15, 15)
                image    = TF.rotate(image,    angle,
                                     interpolation=Image.BILINEAR)
                mask_pil = TF.rotate(mask_pil, angle,
                                     interpolation=Image.NEAREST)

            # Random crop + resize (scale 0.7–1.0)
            if random.random() > 0.4:
                scale  = random.uniform(0.7, 1.0)
                crop_h = int(self.h * scale)
                crop_w = int(self.w * scale)
                max_i  = max(self.h - crop_h, 1)
                max_j  = max(self.w - crop_w, 1)
                i = random.randint(0, max_i)
                j = random.randint(0, max_j)
                image    = TF.crop(image,    i, j, crop_h, crop_w)
                mask_pil = TF.crop(mask_pil, i, j, crop_h, crop_w)
                image    = TF.resize(image,
                                     (self.h, self.w), interpolation=Image.BILINEAR)
                mask_pil = TF.resize(mask_pil,
                                     (self.h, self.w), interpolation=Image.NEAREST)

            # Colour jitter 70%
            if random.random() > 0.3:
                image = TF.adjust_brightness(image, random.uniform(0.7, 1.3))
                image = TF.adjust_contrast  (image, random.uniform(0.7, 1.3))
                image = TF.adjust_saturation(image, random.uniform(0.7, 1.3))
                image = TF.adjust_hue       (image, random.uniform(-0.08, 0.08))

        image = TF.to_tensor(image)
        image = TF.normalize(image,
                             mean=[0.485, 0.456, 0.406],
                             std =[0.229, 0.224, 0.225])
        mask  = torch.from_numpy(np.array(mask_pil)).long()
        return image, mask


# ─────────────────────────────────────────────
#  DATASET  — only loads matched image+mask pairs
# ─────────────────────────────────────────────
class SegDataset(Dataset):
    def __init__(self, data_dir: str, augment: JointAugment):
        self.image_dir = os.path.join(data_dir, 'Color_Images')
        self.masks_dir = os.path.join(data_dir, 'Segmentation')
        self.augment   = augment

        all_images = sorted(os.listdir(self.image_dir))
        self.ids   = [
            f for f in all_images
            if os.path.isfile(os.path.join(self.image_dir, f))          # ← folder skip karo
            and f.endswith(('.png', '.jpg', '.jpeg'))                    # ← sirf images lo
            and os.path.isfile(os.path.join(self.masks_dir, f))         # ← mask bhi hona chahiye
        ]

        skipped = len(all_images) - len(self.ids)
        if skipped > 0:
            print(f"  ⚠ Skipped {skipped} items — folders/no mask")
        print(f"  ✓ Matched image-mask pairs: {len(self.ids)}")

    def __len__(self):
        return len(self.ids)

    def __getitem__(self, idx):
        name    = self.ids[idx]
        image   = Image.open(os.path.join(self.image_dir, name)).convert('RGB')
        mask    = Image.open(os.path.join(self.masks_dir, name))
        mask_np = convert_mask(mask)
        image, mask = self.augment(image, mask_np)
        return image, mask


# ─────────────────────────────────────────────
#  METRICS
# ─────────────────────────────────────────────
def compute_iou(logits, target, num_classes=N_CLASSES):
    pred = logits.argmax(dim=1).view(-1)
    tgt  = target.view(-1)
    per  = []
    for c in range(num_classes):
        p = pred == c;  t = tgt == c
        inter = (p & t).sum().float()
        union = (p | t).sum().float()
        per.append(float('nan') if union == 0
                   else (inter / union).item())
    return float(np.nanmean(per)), per


def compute_dice(logits, target, num_classes=N_CLASSES, smooth=1e-6):
    pred = logits.argmax(dim=1).view(-1)
    tgt  = target.view(-1)
    sc   = []
    for c in range(num_classes):
        p = (pred == c).float();  t = (tgt == c).float()
        sc.append(((2*(p*t).sum() + smooth) /
                   (p.sum() + t.sum() + smooth)).item())
    return float(np.mean(sc))


def compute_pixel_acc(logits, target):
    return (logits.argmax(dim=1) == target).float().mean().item()


@torch.no_grad()
def evaluate(model, loader, device, criterion):
    model.eval()
    iou_l, dice_l, acc_l, loss_l = [], [], [], []
    for imgs, labels in loader:
        imgs, labels = imgs.to(device), labels.to(device)
        with torch.amp.autocast('cuda'):
            out    = model(pixel_values=imgs)
            logits = F.interpolate(out.logits, size=imgs.shape[2:],
                                   mode='bilinear', align_corners=False)
            loss   = criterion(logits, labels)
        iou, _ = compute_iou(logits, labels)
        iou_l.append(iou)
        dice_l.append(compute_dice(logits, labels))
        acc_l.append(compute_pixel_acc(logits, labels))
        loss_l.append(loss.item())
    model.train()
    return (float(np.nanmean(iou_l)), float(np.mean(dice_l)),
            float(np.mean(acc_l)),    float(np.mean(loss_l)))


# ─────────────────────────────────────────────
#  PLOTS + LOG
# ─────────────────────────────────────────────
def save_plots(history, out_dir):
    os.makedirs(out_dir, exist_ok=True)
    epochs = range(1, len(history['train_loss']) + 1)
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    pairs = [
        ('train_loss','val_loss','Loss','Loss'),
        ('train_iou', 'val_iou', 'IoU', 'IoU'),
        ('train_dice','val_dice','Dice Score','Dice'),
        ('train_pixel_acc','val_pixel_acc','Pixel Accuracy','Accuracy'),
    ]
    for ax, (tk, vk, title, ylabel) in zip(axes.flat, pairs):
        ax.plot(epochs, history[tk], 'o-', label='Train', linewidth=2)
        ax.plot(epochs, history[vk], 's-', label='Val',   linewidth=2)
        ax.set_title(title, fontsize=13, fontweight='bold')
        ax.set_xlabel('Epoch'); ax.set_ylabel(ylabel)
        ax.legend(); ax.grid(True, alpha=0.3)
    plt.tight_layout()
    p = os.path.join(out_dir, 'all_metrics.png')
    plt.savefig(p, dpi=150); plt.close()
    print(f"  Saved → {p}")


def save_per_class_plot(per_class, out_dir, epoch):
    os.makedirs(out_dir, exist_ok=True)
    vals   = [x if not np.isnan(x) else 0 for x in per_class]
    colors = ['#2ecc71' if v >= 0.5 else '#e67e22' if v >= 0.3
              else '#e74c3c' for v in vals]
    fig, ax = plt.subplots(figsize=(12, 5))
    ax.bar(CLASS_NAMES, vals, color=colors, edgecolor='white')
    mean_v = float(np.nanmean(per_class))
    ax.axhline(mean_v, color='red', linestyle='--', linewidth=1.5,
               label=f'Mean IoU = {mean_v:.3f}')
    ax.set_ylim(0, 1); ax.set_ylabel('IoU')
    ax.set_title(f'Per-Class IoU — Epoch {epoch}', fontweight='bold')
    ax.legend(); ax.grid(axis='y', alpha=0.3)
    plt.xticks(rotation=35, ha='right'); plt.tight_layout()
    p = os.path.join(out_dir, f'per_class_iou_epoch{epoch:02d}.png')
    plt.savefig(p, dpi=150); plt.close()
    print(f"  Saved → {p}")


def save_log(history, best_epoch, out_dir):
    os.makedirs(out_dir, exist_ok=True)
    p = os.path.join(out_dir, 'training_log.txt')
    with open(p, 'w') as f:
        f.write("=" * 80 + "\n")
        f.write("  DUALITY AI HACKATHON — SEGFORMER-B2 TRAINING LOG\n")
        f.write("=" * 80 + "\n\n")
        f.write(f"  Model         : {CONFIG['model_name']}\n")
        f.write(f"  Image size    : {CONFIG['img_h']} x {CONFIG['img_w']}\n")
        f.write(f"  Batch size    : {CONFIG['batch_size']}\n")
        f.write(f"  Learning rate : {CONFIG['lr']}\n\n")
        f.write(f"  Best Val IoU  : {max(history['val_iou']):.4f}  (Epoch {best_epoch})\n")
        f.write(f"  Best Val Dice : {max(history['val_dice']):.4f}\n")
        f.write(f"  Best Val Acc  : {max(history['val_pixel_acc']):.4f}\n\n")
        hdr = ('Epoch','TrLoss','VaLoss','TrIoU','VaIoU',
               'TrDice','VaDice','TrAcc','VaAcc')
        f.write("{:<7}{:<10}{:<10}{:<10}{:<10}"
                "{:<10}{:<10}{:<10}{:<10}\n".format(*hdr))
        f.write("-" * 80 + "\n")
        for i in range(len(history['train_loss'])):
            mark = " ★" if (i + 1) == best_epoch else ""
            f.write("{:<7}{:<10.4f}{:<10.4f}{:<10.4f}{:<10.4f}"
                    "{:<10.4f}{:<10.4f}{:<10.4f}{:<10.4f}{}\n".format(
                i+1,
                history['train_loss'][i], history['val_loss'][i],
                history['train_iou'][i],  history['val_iou'][i],
                history['train_dice'][i], history['val_dice'][i],
                history['train_pixel_acc'][i], history['val_pixel_acc'][i],
                mark))
    print(f"  Saved → {p}")


# ─────────────────────────────────────────────
#  MAIN
# ─────────────────────────────────────────────
def main():
    # ── Paths ─────────────────────────────────
    # 🔁 Change base_dir if your Drive folder is different
    base_dir  = "/content/drive/MyDrive/Offroad_Segmentation_Training_Dataset/Offroad_Segmentation_Training_Dataset"
    train_dir = os.path.join(base_dir, "train")
    val_dir   = os.path.join(base_dir, "val")
    out_dir   = os.path.join(base_dir, "outputs")
    last_path = os.path.join(base_dir, "segformer_last.pth")
    best_path = os.path.join(base_dir, "segformer_best.pth")
    os.makedirs(out_dir, exist_ok=True)

    # ── Device ────────────────────────────────
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"\n{'='*60}")
    print(f"  Device        : {device}")
    if device.type == 'cuda':
        print(f"  GPU           : {torch.cuda.get_device_name(0)}")
        vram = torch.cuda.get_device_properties(0).total_memory / 1e9
        print(f"  VRAM          : {vram:.1f} GB")
    print(f"  Model         : {CONFIG['model_name']}")
    print(f"  Image size    : {CONFIG['img_h']} x {CONFIG['img_w']}")
    print(f"  Batch size    : {CONFIG['batch_size']}")
    print(f"  Epochs        : {CONFIG['n_epochs']}")
    print(f"  LR            : {CONFIG['lr']}")
    print(f"{'='*60}\n")

    # ── Datasets ──────────────────────────────
    print("Loading datasets ...")
    train_aug = JointAugment(CONFIG['img_h'], CONFIG['img_w'], train=True)
    val_aug   = JointAugment(CONFIG['img_h'], CONFIG['img_w'], train=False)

    print("  Train set:")
    train_set = SegDataset(train_dir, train_aug)
    print("  Val set:")
    val_set   = SegDataset(val_dir, val_aug)

    # num_workers=2 for Colab (4 causes warnings on T4)
    train_loader = DataLoader(train_set,
                              batch_size=CONFIG['batch_size'],
                              shuffle=True,
                              num_workers=2,
                              pin_memory=True,
                              prefetch_factor=2)
    val_loader   = DataLoader(val_set,
                              batch_size=CONFIG['batch_size'],
                              shuffle=False,
                              num_workers=2,
                              pin_memory=True,
                              prefetch_factor=2)

    print(f"\n  Train samples : {len(train_set)}")
    print(f"  Val   samples : {len(val_set)}\n")

    # ── Load SegFormer-B2 ──────────────────────
    print(f"  Loading {CONFIG['model_name']} ...")
    print("  (First run downloads ~350 MB — cached after that)\n")
    model = SegformerForSemanticSegmentation.from_pretrained(
        CONFIG['model_name'],
        num_labels=N_CLASSES,
        ignore_mismatched_sizes=True,
        id2label={i: n for i, n in enumerate(CLASS_NAMES)},
        label2id={n: i for i, n in enumerate(CLASS_NAMES)},
    ).to(device)

    tp = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"  Trainable params : {tp:,}\n")

    # ── Loss (weighted) ───────────────────────
    weights   = torch.tensor(CONFIG['class_weights'], dtype=torch.float).to(device)
    criterion = nn.CrossEntropyLoss(weight=weights)

    # ── Optimiser — encoder 10x smaller LR ───
    optimizer = optim.AdamW([
        {'params': model.segformer.parameters(),   'lr': CONFIG['lr'] * 0.1},
        {'params': model.decode_head.parameters(), 'lr': CONFIG['lr']},
    ], weight_decay=CONFIG['weight_decay'])

    scheduler = optim.lr_scheduler.CosineAnnealingLR(
        optimizer, T_max=CONFIG['n_epochs'], eta_min=1e-7)

    # ── AMP — use new API to avoid deprecation warning ──
    use_amp = device.type == 'cuda'
    scaler  = torch.amp.GradScaler('cuda') if use_amp else None

    # ── History ───────────────────────────────
    history = {k: [] for k in [
        'train_loss','val_loss','train_iou','val_iou',
        'train_dice','val_dice','train_pixel_acc','val_pixel_acc']}

    best_val_iou, best_epoch, patience_count = 0.0, 1, 0

    print("=" * 60)
    print("  TRAINING START")
    print("=" * 60)

    for epoch in range(1, CONFIG['n_epochs'] + 1):
        model.train()
        train_losses = []
        pbar = tqdm(train_loader,
                    desc=f"Epoch {epoch:02d}/{CONFIG['n_epochs']} [Train]",
                    leave=False)

        for imgs, labels in pbar:
            imgs, labels = imgs.to(device), labels.to(device)
            optimizer.zero_grad(set_to_none=True)

            if use_amp:
                # ── use new torch.amp.autocast API ──
                with torch.amp.autocast('cuda'):
                    out    = model(pixel_values=imgs)
                    logits = F.interpolate(out.logits, size=imgs.shape[2:],
                                           mode='bilinear', align_corners=False)
                    loss   = criterion(logits, labels)
                scaler.scale(loss).backward()
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                scaler.step(optimizer)
                scaler.update()
            else:
                out    = model(pixel_values=imgs)
                logits = F.interpolate(out.logits, size=imgs.shape[2:],
                                       mode='bilinear', align_corners=False)
                loss   = criterion(logits, labels)
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()

            train_losses.append(loss.item())
            pbar.set_postfix(loss=f"{loss.item():.4f}")

        scheduler.step()

        # ── Validation metrics ────────────────
        val_iou,   val_dice,   val_acc,   val_loss   = evaluate(
            model, val_loader,   device, criterion)
        train_iou, train_dice, train_acc, _           = evaluate(
            model, train_loader, device, criterion)
        t_loss = float(np.mean(train_losses))

        for k, v in zip(
            ['train_loss','val_loss','train_iou','val_iou',
             'train_dice','val_dice','train_pixel_acc','val_pixel_acc'],
            [t_loss, val_loss, train_iou, val_iou,
             train_dice, val_dice, train_acc, val_acc]
        ):
            history[k].append(v)

        # ── Per-class IoU (val set) ───────────
        model.eval()
        accum = [[] for _ in range(N_CLASSES)]
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs, labels = imgs.to(device), labels.to(device)
                with torch.amp.autocast('cuda'):
                    out    = model(pixel_values=imgs)
                    logits = F.interpolate(out.logits, size=imgs.shape[2:],
                                           mode='bilinear', align_corners=False)
                _, pc = compute_iou(logits, labels)
                for c in range(N_CLASSES):
                    if not np.isnan(pc[c]):
                        accum[c].append(pc[c])
        avg_pc = [float(np.mean(v)) if v else float('nan') for v in accum]
        model.train()

        # ── Console output ────────────────────
        lr_d = optimizer.param_groups[1]['lr']
        print(f"\nEpoch {epoch:02d}/{CONFIG['n_epochs']}  lr={lr_d:.1e}  |  "
              f"T-loss={t_loss:.4f}  V-loss={val_loss:.4f}  |  "
              f"V-IoU={val_iou:.4f}  V-Dice={val_dice:.4f}  V-Acc={val_acc:.4f}")
        print("  Per-class IoU:")
        for name, iou in zip(CLASS_NAMES, avg_pc):
            flag = "  ← LOW" if (not np.isnan(iou) and iou < 0.3) else ""
            bar  = "█" * int((iou if not np.isnan(iou) else 0) * 20)
            print(f"    {name:<20} {iou:.3f}  {bar}{flag}")

        # Save per-class plot every 5 epochs
        if epoch % 5 == 0 or epoch == CONFIG['n_epochs']:
            save_per_class_plot(avg_pc, out_dir, epoch)

        # ── Save best checkpoint ──────────────
        if val_iou > best_val_iou:
            best_val_iou   = val_iou
            best_epoch     = epoch
            patience_count = 0
            torch.save(model.state_dict(), best_path)
            print(f"  ★  New best Val IoU = {best_val_iou:.4f}  → saved to Drive")
        else:
            patience_count += 1
            print(f"  No improvement ({patience_count}/{CONFIG['patience']})")

        # ── Early stopping ────────────────────
        if patience_count >= CONFIG['patience']:
            print(f"\n  Early stopping triggered at epoch {epoch}.")
            break

        # ── Free VRAM between epochs ──────────
        torch.cuda.empty_cache()

    # ── Save last model + plots + log ─────────
    torch.save(model.state_dict(), last_path)
    print(f"\n  Saved last model  → {last_path}")
    print(f"  Saved best model  → {best_path}")

    save_plots(history, out_dir)
    save_log(history, best_epoch, out_dir)

    print("\n" + "=" * 60)
    print("  TRAINING COMPLETE")
    print("=" * 60)
    print(f"  Best Val IoU   : {best_val_iou:.4f}  (Epoch {best_epoch})")
    print(f"  Best Val Dice  : {max(history['val_dice']):.4f}")
    print(f"  Best Val Acc   : {max(history['val_pixel_acc']):.4f}")
    print(f"  Outputs saved  : {out_dir}")
    print("=" * 60)


# ─────────────────────────────────────────────
#  RUN
# ─────────────────────────────────────────────
if __name__ == "__main__":
    main()

# For Colab — also works when called directly (not as __main__)
# Just call main() if running as a notebook cell
```

```python
"""
=============================================================================
  Duality AI Hackathon — Val/Inference Script (Google Colab)
  Loads best trained SegFormer-B2 model and runs on test images

  Before running:
      from google.colab import drive
      drive.mount('/content/drive')
=============================================================================
"""

import os
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import torch
import torch.nn.functional as F
import torchvision.transforms.functional as TF
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import cv2
from tqdm import tqdm
from transformers import SegformerForSemanticSegmentation


# ─────────────────────────────────────────────
#  PATHS — change these if needed
# ─────────────────────────────────────────────
BASE_DIR   = "/content/drive/MyDrive/Offroad_Segmentation_Training_Dataset/Offroad_Segmentation_Training_Dataset"
MODEL_PATH = os.path.join(BASE_DIR, "segformer_best.pth")
DATA_DIR   = os.path.join("/content/drive/MyDrive/Offroad_Segmentation_testImages/Offroad_Segmentation_testImages")   # 🔁 change if different
OUT_DIR    = os.path.join(BASE_DIR, "predictions")
MODEL_NAME = "nvidia/mit-b2"

IMG_H, IMG_W = 512, 512
BATCH_SIZE   = 4
NUM_COMPARE  = 15   # kitne side-by-side images save karo report ke liye


# ─────────────────────────────────────────────
#  CLASS DEFINITIONS
# ─────────────────────────────────────────────
VALUE_MAP = {
    0:0, 100:1, 200:2, 300:3, 500:4,
    550:5, 700:6, 800:7, 7100:8, 10000:9
}
N_CLASSES   = 10
CLASS_NAMES = [
    'Background', 'Trees', 'Lush Bushes', 'Dry Grass', 'Dry Bushes',
    'Ground Clutter', 'Logs', 'Rocks', 'Landscape', 'Sky'
]
COLOR_PALETTE = np.array([
    [ 30,  30,  30],   # Background
    [ 34, 139,  34],   # Trees
    [  0, 220,  80],   # Lush Bushes
    [210, 180, 100],   # Dry Grass
    [160, 100,  40],   # Dry Bushes
    [180, 180,   0],   # Ground Clutter
    [139,  60,  19],   # Logs
    [160, 160, 160],   # Rocks
    [200, 100,  50],   # Landscape
    [100, 180, 240],   # Sky
], dtype=np.uint8)


# ─────────────────────────────────────────────
#  HELPERS
# ─────────────────────────────────────────────
def convert_mask(mask):
    arr = np.array(mask)
    out = np.zeros_like(arr, dtype=np.uint8)
    for raw, cls in VALUE_MAP.items():
        out[arr == raw] = cls
    return out


def mask_to_color(mask_np):
    canvas = np.zeros((*mask_np.shape, 3), dtype=np.uint8)
    for c in range(N_CLASSES):
        canvas[mask_np == c] = COLOR_PALETTE[c]
    return canvas


# ─────────────────────────────────────────────
#  DATASET
#  Works for both:
#    - Structured: Color_Images/ + Segmentation/
#    - Flat: images directly in folder
# ─────────────────────────────────────────────
class TestDataset(Dataset):
    def __init__(self, data_dir, img_h, img_w):
        self.h, self.w = img_h, img_w

        color_dir = os.path.join(data_dir, 'Color_Images')
        seg_dir   = os.path.join(data_dir, 'Segmentation')

        # Auto-detect folder structure
        self.image_dir = color_dir if os.path.isdir(color_dir) else data_dir
        self.masks_dir = seg_dir   if os.path.isdir(seg_dir)   else None

        self.fnames = sorted([
            f for f in os.listdir(self.image_dir)
            if os.path.isfile(os.path.join(self.image_dir, f))
            and f.lower().endswith(('.png', '.jpg', '.jpeg'))
        ])
        print(f"  ✓ Found {len(self.fnames)} test images")
        if self.masks_dir:
            print(f"  ✓ Ground truth masks found — will compute IoU")
        else:
            print(f"  ℹ No masks folder — prediction only mode")

    def __len__(self):
        return len(self.fnames)

    def __getitem__(self, idx):
        name  = self.fnames[idx]
        image = Image.open(
            os.path.join(self.image_dir, name)).convert('RGB')
        image = TF.resize(image, (self.h, self.w),
                          interpolation=Image.BILINEAR)
        image = TF.to_tensor(image)
        image = TF.normalize(image,
                             mean=[0.485, 0.456, 0.406],
                             std =[0.229, 0.224, 0.225])

        # Load mask if exists
        if self.masks_dir:
            mp = os.path.join(self.masks_dir, name)
            if os.path.isfile(mp):
                m    = Image.open(mp)
                mnp  = convert_mask(m)
                mpil = Image.fromarray(mnp)
                mpil = TF.resize(mpil, (self.h, self.w),
                                 interpolation=Image.NEAREST)
                return image, torch.from_numpy(
                    np.array(mpil)).long(), name

        return image, torch.full(
            (self.h, self.w), -1, dtype=torch.long), name


# ─────────────────────────────────────────────
#  METRICS
# ─────────────────────────────────────────────
def iou_per_class(pred_np, target_np):
    results = []
    for c in range(N_CLASSES):
        p     = pred_np   == c
        t     = target_np == c
        inter = np.logical_and(p, t).sum()
        union = np.logical_or(p, t).sum()
        results.append(float('nan') if union == 0 else inter / union)
    return results


def conf_matrix(pred_np, target_np):
    valid = target_np >= 0
    p  = pred_np[valid].astype(np.int32)
    t  = target_np[valid].astype(np.int32)
    cm = np.zeros((N_CLASSES, N_CLASSES), dtype=np.int64)
    for pi, ti in zip(p, t):
        if 0 <= ti < N_CLASSES and 0 <= pi < N_CLASSES:
            cm[ti, pi] += 1
    return cm


# ─────────────────────────────────────────────
#  SAVE FUNCTIONS
# ─────────────────────────────────────────────
def save_comparison(img_t, gt_t, pred_t, path, title=''):
    mean = np.array([0.485, 0.456, 0.406])
    std  = np.array([0.229, 0.224, 0.225])
    img  = np.clip(
        img_t.cpu().numpy().transpose(1, 2, 0) * std + mean, 0, 1)

    has_gt = (gt_t >= 0).any().item()
    ncols  = 3 if has_gt else 2
    fig, axes = plt.subplots(1, ncols, figsize=(6 * ncols, 5))

    axes[0].imshow(img)
    axes[0].set_title('Input Image'); axes[0].axis('off')

    if has_gt:
        axes[1].imshow(mask_to_color(
            gt_t.cpu().numpy().astype(np.uint8)))
        axes[1].set_title('Ground Truth'); axes[1].axis('off')
        axes[2].imshow(mask_to_color(
            pred_t.cpu().numpy().astype(np.uint8)))
        axes[2].set_title('Prediction'); axes[2].axis('off')
    else:
        axes[1].imshow(mask_to_color(
            pred_t.cpu().numpy().astype(np.uint8)))
        axes[1].set_title('Prediction'); axes[1].axis('off')

    patches = [
        mpatches.Patch(color=COLOR_PALETTE[c]/255, label=CLASS_NAMES[c])
        for c in range(N_CLASSES)
    ]
    fig.legend(handles=patches, loc='lower center', ncol=5,
               fontsize=8, bbox_to_anchor=(0.5, -0.05))
    if title:
        fig.suptitle(title, fontsize=10)
    plt.tight_layout()
    plt.savefig(path, dpi=150, bbox_inches='tight')
    plt.close()


def save_iou_bar(mean_iou, per_class, path):
    vals   = [v if not np.isnan(v) else 0 for v in per_class]
    colors = [COLOR_PALETTE[c] / 255 for c in range(N_CLASSES)]
    fig, ax = plt.subplots(figsize=(12, 5))
    bars = ax.bar(CLASS_NAMES, vals, color=colors, edgecolor='white')
    for bar, val in zip(bars, vals):
        ax.text(bar.get_x() + bar.get_width()/2,
                bar.get_height() + 0.01,
                f'{val:.3f}', ha='center', va='bottom', fontsize=9)
    ax.axhline(mean_iou, color='red', linestyle='--', linewidth=1.5,
               label=f'Mean IoU = {mean_iou:.4f}')
    ax.set_ylim(0, 1.1); ax.set_ylabel('IoU', fontsize=12)
    ax.set_title('Per-Class IoU — Test Set',
                 fontsize=13, fontweight='bold')
    ax.legend(fontsize=11); ax.grid(axis='y', alpha=0.3)
    plt.xticks(rotation=35, ha='right', fontsize=10)
    plt.tight_layout()
    plt.savefig(path, dpi=150, bbox_inches='tight')
    plt.close()
    print(f"  Saved → {path}")


def save_conf_matrix(cm, path):
    row_sum = cm.sum(axis=1, keepdims=True)
    cm_norm = np.divide(cm.astype(float), row_sum,
                        out=np.zeros_like(cm, dtype=float),
                        where=row_sum != 0)
    fig, ax = plt.subplots(figsize=(10, 8))
    im = ax.imshow(cm_norm, cmap='Blues', vmin=0, vmax=1)
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    ticks = np.arange(N_CLASSES)
    ax.set_xticks(ticks)
    ax.set_xticklabels(CLASS_NAMES, rotation=45, ha='right', fontsize=9)
    ax.set_yticks(ticks)
    ax.set_yticklabels(CLASS_NAMES, fontsize=9)
    for i in range(N_CLASSES):
        for j in range(N_CLASSES):
            v = cm_norm[i, j]
            ax.text(j, i, f'{v:.2f}', ha='center', va='center',
                    fontsize=7, color='white' if v > 0.6 else 'black')
    ax.set_xlabel('Predicted', fontsize=11)
    ax.set_ylabel('Ground Truth', fontsize=11)
    ax.set_title('Normalised Confusion Matrix',
                 fontsize=12, fontweight='bold')
    plt.tight_layout()
    plt.savefig(path, dpi=150, bbox_inches='tight')
    plt.close()
    print(f"  Saved → {path}")


def save_metrics_txt(mean_iou, per_class, n_images, path):
    with open(path, 'w') as f:
        f.write("=" * 60 + "\n")
        f.write("  DUALITY AI HACKATHON — EVALUATION RESULTS\n")
        f.write("  Model: SegFormer-B2  (nvidia/mit-b2)\n")
        f.write("=" * 60 + "\n\n")
        f.write(f"  Images evaluated : {n_images}\n")
        f.write(f"  Mean IoU         : {mean_iou:.4f}\n\n")
        f.write("-" * 60 + "\n")
        f.write("  Per-Class IoU\n")
        f.write("-" * 60 + "\n")
        for name, iou in zip(CLASS_NAMES, per_class):
            bar     = '█' * int((iou if not np.isnan(iou) else 0) * 20)
            flag    = '  ← needs improvement' if (
                not np.isnan(iou) and iou < 0.3) else ''
            iou_str = f"{iou:.4f}" if not np.isnan(iou) else "N/A   "
            f.write(f"  {name:<20}  {iou_str}  {bar}{flag}\n")
        f.write("\n" + "=" * 60 + "\n")
    print(f"  Saved → {path}")


# ─────────────────────────────────────────────
#  MAIN INFERENCE
# ─────────────────────────────────────────────
def run_inference():
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    print(f"\n{'='*60}")
    print(f"  INFERENCE — SegFormer-B2")
    print(f"{'='*60}")
    print(f"  Device : {device}")
    print(f"  Model  : {MODEL_PATH}")
    print(f"  Data   : {DATA_DIR}")
    print(f"  Output : {OUT_DIR}")
    print(f"{'='*60}\n")

    # ── Check model exists ────────────────────
    if not os.path.exists(MODEL_PATH):
        fallback = os.path.join(BASE_DIR, "segformer_last.pth")
        if os.path.exists(fallback):
            print(f"  ⚠ Best model not found — using last model")
            model_path = fallback
        else:
            print("  ✗ No model found! Run training first.")
            return
    else:
        model_path = MODEL_PATH
        size_mb = os.path.getsize(model_path) / 1e6
        print(f"  ✓ Model found  ({size_mb:.0f} MB)\n")

    # ── Dataset ───────────────────────────────
    dataset = TestDataset(DATA_DIR, IMG_H, IMG_W)
    loader  = DataLoader(dataset,
                         batch_size=BATCH_SIZE,
                         shuffle=False,
                         num_workers=2,
                         pin_memory=True)

    # ── Load model ────────────────────────────
    print(f"\n  Loading model architecture ...")
    model = SegformerForSemanticSegmentation.from_pretrained(
        MODEL_NAME,
        num_labels=N_CLASSES,
        ignore_mismatched_sizes=True,
        id2label={i: n for i, n in enumerate(CLASS_NAMES)},
        label2id={n: i for i, n in enumerate(CLASS_NAMES)},
    )
    model.load_state_dict(
        torch.load(model_path, map_location=device))
    model.to(device).eval()
    print("  ✓ Model loaded successfully!\n")

    # ── Output directories ────────────────────
    masks_dir   = os.path.join(OUT_DIR, 'masks')
    color_dir   = os.path.join(OUT_DIR, 'masks_color')
    compare_dir = os.path.join(OUT_DIR, 'comparisons')
    for d in [masks_dir, color_dir, compare_dir]:
        os.makedirs(d, exist_ok=True)

    # ── Inference loop ────────────────────────
    print("  Running inference ...\n")
    accum_iou = [[] for _ in range(N_CLASSES)]
    cm_total  = np.zeros((N_CLASSES, N_CLASSES), dtype=np.int64)
    n_saved   = 0
    has_gt    = False

    pbar = tqdm(loader, desc="Processing", unit="batch")

    with torch.no_grad():
        for imgs, labels, names in pbar:
            imgs   = imgs.to(device)
            labels = labels.to(device)

            with torch.amp.autocast('cuda'):
                out    = model(pixel_values=imgs)
                logits = F.interpolate(
                    out.logits, size=(IMG_H, IMG_W),
                    mode='bilinear', align_corners=False)

            preds = logits.argmax(dim=1)

            for i in range(imgs.shape[0]):
                name     = names[i]
                base     = os.path.splitext(name)[0]
                pred_np  = preds[i].cpu().numpy().astype(np.uint8)
                label_np = labels[i].cpu().numpy()

                # Save raw mask (class IDs 0-9) — SUBMIT THIS
                Image.fromarray(pred_np).save(
                    os.path.join(masks_dir, f'{base}_pred.png'))

                # Save colour mask
                cv2.imwrite(
                    os.path.join(color_dir,
                                 f'{base}_pred_color.png'),
                    cv2.cvtColor(mask_to_color(pred_np),
                                 cv2.COLOR_RGB2BGR))

                # Compute metrics if GT available
                if (label_np >= 0).any():
                    has_gt = True
                    pc = iou_per_class(pred_np, label_np)
                    for c in range(N_CLASSES):
                        if not np.isnan(pc[c]):
                            accum_iou[c].append(pc[c])
                    cm_total += conf_matrix(pred_np, label_np)

                # Save comparison images for report
                if n_saved < NUM_COMPARE:
                    save_comparison(
                        imgs[i], labels[i], preds[i],
                        os.path.join(compare_dir,
                                     f'compare_{n_saved:03d}_{base}.png'),
                        title=name)

                n_saved += 1

            # Show running IoU in progress bar
            if has_gt and any(accum_iou):
                run_iou = float(np.nanmean([
                    np.mean(v) if v else float('nan')
                    for v in accum_iou]))
                pbar.set_postfix(mIoU=f"{run_iou:.4f}")

    # ── Results ───────────────────────────────
    print(f"\n  Processed {n_saved} images\n")

    if has_gt:
        avg_pc   = [float(np.mean(v)) if v else float('nan')
                    for v in accum_iou]
        mean_iou = float(np.nanmean(avg_pc))

        print("=" * 60)
        print("  FINAL EVALUATION RESULTS")
        print("=" * 60)
        print(f"  Mean IoU : {mean_iou:.4f}\n")
        print(f"  {'Class':<22} {'IoU':>6}  Progress")
        print("  " + "-" * 50)
        for name, iou in zip(CLASS_NAMES, avg_pc):
            bar  = '█' * int((iou if not np.isnan(iou) else 0) * 20)
            flag = '  ← LOW' if (not np.isnan(iou) and iou < 0.3) else ''
            print(f"  {name:<22} {iou:.4f}  {bar}{flag}")
        print("=" * 60)

        # Save all report files
        save_iou_bar(mean_iou, avg_pc,
                     os.path.join(OUT_DIR, 'per_class_iou.png'))
        save_conf_matrix(cm_total,
                         os.path.join(OUT_DIR, 'confusion_matrix.png'))
        save_metrics_txt(mean_iou, avg_pc, n_saved,
                         os.path.join(OUT_DIR, 'evaluation_metrics.txt'))
    else:
        print("  ℹ No ground truth masks — predictions saved only")
        print("  Submit the masks/ folder")

    # ── Final summary ─────────────────────────
    print(f"\n{'='*60}")
    print(f"  ALL OUTPUTS SAVED TO:")
    print(f"  {OUT_DIR}/")
    print(f"{'='*60}")
    print(f"  masks/               ← SUBMIT THESE to organizers")
    print(f"  masks_color/         ← coloured visuals")
    print(f"  comparisons/         ← {min(n_saved, NUM_COMPARE)} examples for report")
    if has_gt:
        print(f"  per_class_iou.png    ← put in report slide")
        print(f"  confusion_matrix.png ← put in report")
        print(f"  evaluation_metrics.txt ← IoU scores")
    print(f"{'='*60}\n")


# ─────────────────────────────────────────────
#  RUN
# ─────────────────────────────────────────────
run_inference()
```
